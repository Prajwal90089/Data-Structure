```python
import matplotlib.pyplot as plt
%matplotlib inline

class search():
  def __init__(self):
    self.linearSearch=[]
    self.binarysearch=[]
    self.x=[1,2]
    self.y=[]
    self.n=0

  def linear(self):
    print("Present Array :",self.linearSearch)
    flag=0
    step=0
    targetSearch=float(input("enter the element to be get searched in the array :"))
    for i in range(0,len(self.linearSearch)):
      step=step+1
      if(self.linearSearch[i]==targetSearch):
        flag=1
    if flag==1:
      print("Element found in the array")
      self.y.append(step)
      print("no of itterations counted",self.y)

    else:
      print("Element not found in the array")

  def binary(self):
    print("present array :",self.binarysearch)
    self.binarysearch.sort()
    print("Sorted Elements :",self.binarysearch)
    targetElement=float(input("enter the element to be get searched in the array :"))
    left=0
    right=self.n-1
    step=0
    while(left<=right):
      step=step+1
      mid=(left+right)//2
      if(self.binarysearch[mid]==targetElement):
        print("element Found in the array")
        self.y.append(step)
        print("No of Steps encountered in the Linear Search and Binary Search")
        plt.bar(self.x,self.y)
        plt.title("Difference Between Lenear And Binary Search")
        plt.xlabel("1st : Linear 2nd: Binary")
        plt.ylabel("No of Steps")
        return
      elif(self.binarysearch[mid]<targetElement):
        left=mid+1
      else:
        right=mid-1



start=search()
start.n=int(input("Enter the number of elements :"))
array=list(map(float,input("Enter the elements seperated by the space :").strip().split()))
if len(array)==start.n:
  for i in range(0,start.n):
    start.binarysearch.append(array[i])
  for i in range(0,start.n):
    start.linearSearch.append(array[i])
  start.linear()
  start.binary()

else:
  print("enter the valid Dataset equelent to the N Value")
```
